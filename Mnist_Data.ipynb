```python
import tensorflow as tf

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

print("Training data shape:", x_train.shape) 
print("Test data shape:", x_test.shape)      
```

```python
x_train = x_train/255
x_test = x_test/255
x_train = x_train.reshape(60000,784)
x_test= x_test.reshape(10000,784)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from tensorflow.keras import Sequential,Loss,optimizers
from tensorflow.keras.layers import Dense
from tensorflow.keras.metrics import Accuracy
model = Sequential([
    Dense(512,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    Dense(10,activation="softmax")

])
model.compile( optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['Accuracy'])
model.fit(x_train,y_train,batch_size=128,epochs=10)

```

```python
model.evaluate(x_test,y_test)
```

```python
from PIL import Image
import numpy as np
def Mnist_prediction(prediction):
    img = Image.open(prediction)
    Input_image = np.array(img)
    img = img.convert("L")
    Input_image = np.array(img)
    Input_image = Input_image/255
    Input_image = Input_image.reshape(1,784)
    predict = model.predict(Input_image)
    Answer = np.argmax(predict[0])
    return Answer
```
